```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv("./out.csv", delimiter=";")
df
```

```python
def make_get_subplot(gene_prefix: str, label:str):
    base = alt.Chart(df).properties(width=1000)
    boxes = base.mark_rect().encode(
        x=alt.X(f"{gene_prefix}_start:Q").title(None), 
        x2=alt.X2(f"{gene_prefix}_end:Q"), 
        y=alt.Y("name:N", scale=alt.Scale(padding=0.2)).sort(None), 
        color=alt.ColorDatum(label)
    )
    labels = base.mark_text(align="center", size=8).encode(
        y=alt.Y("name:N").sort(None),
        x="tx:Q",
        text=alt.TextDatum(label),
        color=alt.ColorValue("white")
    ).transform_calculate(
        tx=f"( datum.{gene_prefix}_start + datum.{gene_prefix}_end ) / 2"
    )

    return boxes + labels
    
alt.theme.enable('powerbi')
base = alt.Chart(df)
plot = base.mark_rect(height=2).encode(
        x="seqlen",
        y=alt.Y("name:N").sort(None), 
        color=alt.ColorValue("black")
    ) + base.mark_text(align="left", dx=10, size=9).encode(
        y=alt.Y("name:N").sort(None),
        x="seqlen",
        text='seqlen'
    )
ps = ["orf", 's', 'e', 'm', 'n']
ls = ["ORF 1a/1b", "S", "E", "M", "N", ]

for p, l in zip(ps, ls):
    pp = make_get_subplot(p, l)
    if plot is None:
        plot = pp
    else:
        plot = plot + pp

plot
```
